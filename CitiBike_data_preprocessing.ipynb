```python
# Dependencies
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
```

```python
# List of file names to be combined 
files23 = ['data/202301.csv','data/202304.csv', 'data/202307.csv']

# Create an empty list to store each dataframe
df23_list = []

# Loop through each file in the list
for file in files23:
    # Read the file into a dataframe
    df23 = pd.read_csv(file)

    # Append the dataframe to the list
    df23_list.append(df23)

# Concatenate all dataframes in the list into one dataframe
df23 = pd.concat(df23_list)

# Show dataframe
df23
```

```python
df23.info()
```

```python
df23[["started_at","ended_at"]] = df23[["started_at","ended_at"]].apply(pd.to_datetime)
```

```python
df23.info()
```

```python
df23["trip_time_min"] = df23.ended_at-df23.started_at
```

```python
df23["trip_time_days"] = (df23.ended_at-df23.started_at).dt.days
```

```python
df23['time_difference_seconds'] = df23["trip_time_min"].dt.total_seconds()
```

```python
df23.sort_values(by=["trip_time_min"])
```

```python
df23.isnull().sum()
```

```python
df_rmvd_60sec= df23[(df23['time_difference_seconds']> 60)] 
```

```python
df_rmvd_60sec.sort_values(by=['time_difference_seconds'])
```

```python
q1_clean = df_rmvd_60sec['time_difference_seconds'].quantile(0.25)
q3_clean = df_rmvd_60sec['time_difference_seconds'].quantile(0.75)
iqr_clean = q3_clean - q1_clean
lower_bound_clean = q1_clean - 1.5 * iqr_clean
upper_bound_clean = q3_clean + 1.5 * iqr_clean

print("Lower Quartile for the trip time in seconds is: " , q1_clean , "which is", q1_clean/60 , "in minutes")
print("Upper Quartile for the trip time in seconds is: " , q3_clean, "which is", q3_clean/60 , "in minutes")
print("Inter-Quartile Range (IQR) for the trip time in seconds is: " , iqr_clean, "which is", iqr_clean/60,  "in minutes")
print("Lower-bound for the trip time in seconds is: " , lower_bound_clean, "which is", lower_bound_clean/60 , "in minutes")
print("Upper_bound for the trip time in seconds is: " , upper_bound_clean, "which is", upper_bound_clean/60 , "in minutes")
```

```python
below_count = len(df_rmvd_60sec[df_rmvd_60sec['time_difference_seconds'] < lower_bound_clean])
above_count = len(df_rmvd_60sec[df_rmvd_60sec['time_difference_seconds'] > upper_bound_clean])

print(f"Number of values below 1.5 * IQR: {below_count}")
print(f"Number of values above 1.5 * IQR: {above_count}")
```
